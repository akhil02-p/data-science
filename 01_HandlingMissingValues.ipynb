```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Age': [20, 25, 30, 35, np.nan,60],
    'Occupation': ['Finance','Doctor','Professor','Developer',np.nan,'Police'],
    'Salary': [25000,30000,10000,50000,np.nan,30000]
})

df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df_ffill = df.copy()
df_ffill['Occupation'] = df_ffill['Occupation'].ffill()
df_ffill
```

```python
df_bfill = df_ffill.copy()
df_bfill['Salary'] = df_bfill['Salary'].bfill()
df_bfill
```

```python
df_drop_columns = df.dropna()
df_drop_columns
```
