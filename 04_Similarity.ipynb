```python
import numpy as np
from scipy.spatial import distance
```

```python
pointA = np.array([2,4,9])
pointB = np.array([5,1,9])

# euclidean distance
euclidean_distance = distance.euclidean(pointA, pointB)
print(euclidean_distance)

# similarity (inverse of distance)
similarity_euclidean = 1 / (1+euclidean_distance)
print(similarity_euclidean)
```

```python
# manhattan distance
manhattan_distance = distance.cityblock(pointA, pointB)
print(manhattan_distance)

# similarity (inverse of distance)
similarity_manhattan = 1 / (1+manhattan_distance)
print(similarity_manhattan)
```

```python
# minowski
minowski_distance = distance.minkowski(pointA, pointB, p=1.5)
print(minowski_distance)
```

```python
import pandas as pd
from scipy.stats import pearsonr,spearmanr

tips = pd.read_csv("tips.csv")
corr,p_value = pearsonr(tips["total_bill"], tips["tip"])
print(corr)
print(p_value)
```

```python
df = pd.DataFrame(
    {
        'X':[1,2,3],
        'Y':[4,5,6]
    }
)
# spearman correlation
corr_value,p_value = spearmanr(df["X"], df["Y"])

print(corr_value,p_value)
```

```python
# hamming distance
def hamming_distance(string1, string2):
  if len(string1) != len(string2):
    raise ValueError("String lengths are not equal")
  distance = 0
  for i in range(len(string1)):
      if string1[i] != string2[i]:
          distance += 1
  return distance

s1 = "nikhilnikhil"
s2 = "sresyasseryo"
print(hamming_distance(s1,s2))
```

```python
# jaccard index
def jaccard_index(str1, str2):
  set1 = set(str1)
  set2 = set(str2)
  intersection = len(set1.intersection(set2))
  union = len(set1.union(set2))
  return intersection / union

set1 = "data science is fun"
set2 = " science is data"
print(jaccard_index(set1,set2))
```

```python
# co-sine similarity

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

s1 = "data science is fun"
s2 = "science makes data useful"

vectorizer = CountVectorizer()
vectors = vectorizer.fit_transform([s1,s2])
cos_sim = cosine_similarity(vectors[0], vectors[1])[0][0]
print(cos_sim)
```

```python
def lcs_length(X, Y):
  m, n = len(X), len(Y)
  dp = [[0] * (n + 1) for _ in range(m + 1)]
  for i in range(m):
    for j in range(n):
      if X[i] == Y[j]:
        dp[i+1][j+1] = dp[i][j] + 1
      else:
        dp[i+1][j+1] = max(dp[i][j+1],dp[i+1][j])
  return dp[m][n]

seq1 = "ABCDEF"
seq2 = "AEBDF"
length = lcs_length(seq1, seq2)
print(f" LCS '{seq1}' and '{seq2}': {length}")
```
